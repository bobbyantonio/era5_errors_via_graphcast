```python
import datetime
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
lat_range = np.arange(-3,18,0.25)
lon_range = np.arange(30,50,0.25)

era5_2016 = xr.open_dataset("era5_2m_temperature_20160118-20160120.nc").sel(latitude=lat_range, longitude=lon_range).sel(valid_time=slice(datetime.datetime(2016,1,18,0), datetime.datetime(2016,1,18,23)))
```

```python
# Impor metar data
metar_df = pd.read_csv('metar_20160118_06.csv', parse_dates=['date'])
metar_grouped = metar_df.groupby('date').mean().reset_index()
```

```python
# Manually populated from ECMWF plots
approx_fg_points = np.array([17,16.5,15.5, 18, 21, 23, 25, 26.5, 25.5, 24.2, 27.8, 26, 25.5, 24.5, 22, 21, 22,20, 19.5, 19.2, 18, 19.2, 17.8, 13.75 ]) + 273
approx_obs_points = np.array([17.5, 18, 16, 16.5, 16.5, 18, 22, 24, 24.8, 25, 27.5, 27.5, 27.5, 27.9, 26.1, 25, 25.5, 23, 21.8, 21.9, 20, 19.8, 19.8, 17.95]) + 273
approx_analysis_points = np.array([18,18,16.2,17.5, 16.8, 18.1, 20, 25.5, 21.5, 25, 27.5, 27.5, 26.5, 27.9, 26, 22.6, 23.4, 23, 21, 21.5, 20, 19.1, 19.5, 18]) + 273


synop_times = era5_2016['valid_time'].values[:24:3]
approx_synop_fg_points = np.array([np.nan, 16, 28, 24, 34.5, np.nan, 19, 17]) + 273
approx_synop_analysis_points = np.array([np.nan,15, 28.5, 24, 37, np.nan, 20.5, 18]) + 273
approx_synop_obs_points = np.array([np.nan,13.5, 28, 25, 37, np.nan, 21, 19]) + 273
```

```python
fig, ax = plt.subplots(1,1)



ax.plot(era5_2016['valid_time'].values, approx_fg_points, color='blue', label='FG')
ax.plot(era5_2016['valid_time'].values, approx_obs_points, color='black', label='Observed')
ax.plot(era5_2016['valid_time'].values, approx_analysis_points, color='red', label='Analysis')
ax.plot(synop_times, approx_synop_fg_points, 'bo', label='SYNOP FG')
ax.plot(synop_times, approx_synop_analysis_points, 'ro', label='SYNOP Analysis')
ax.plot(synop_times, approx_synop_obs_points, 'ko', label='SYNOP Obs')

ax.plot(metar_grouped['date'], metar_grouped['air_temp'], 'k^', label='METAR')

era5_2016.mean(['latitude','longitude']).t2m.plot(ax=ax, label='ERA5')

plt.legend()
ax.set_title('2m Temperature comparison')
ax.set_ylabel('Temperature (K)')
```
